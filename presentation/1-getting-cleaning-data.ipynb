```python
import requests

url = "https://jsearch.p.rapidapi.com/search"

querystring = {"query": "software engineer intern", "page": "1", "num_pages": "10"}

headers = {
    "X-RapidAPI-Key": API_KEY,
    "X-RapidAPI-Host": "jsearch.p.rapidapi.com",
}

response = requests.get(url, headers=headers, params=querystring)
print(response.json())
```

```python
jobs_data = response.json()["data"]
# clean up nested json
for i, job in enumerate(jobs_data):
    job["job_qualifications"] = ""
    if "Qualifications" in job["job_highlights"]:
        job["job_qualifications"] = " ".join(job["job_highlights"]["Qualifications"])
    job["job_responsibilities"] = ""
    if "Responsibilities" in job["job_highlights"]:
        job["job_responsibilities"] = " ".join(job["job_highlights"]["Responsibilities"])
    del job["job_highlights"]
    del job["apply_options"]
jobs_data
```

```python
# convert to dataframe and set job id as the index
import pandas as pd

job_df = pd.DataFrame(jobs_data)
job_df.set_index("job_id", inplace=True)
job_df
```

```python
job_df["job_employment_type"] = job_df["job_employment_type"].str.title()
job_df["job_full_description"] = (
    job_df["job_description"]
    + " "
    + job_df["job_qualifications"]
    + " "
    + job_df["job_responsibilities"]
)
job_df
```

```python
# this is a function to get the average swe intern salary based on location
# also a cache for performance
cache = {}

def scrape_salary(city, state):
    if (city, state) in cache:
        return cache[(city, state)]
    headers = {
        "accept": "application/json",
        "accept-language": "en-US,en;q=0.9",
        "origin": "https://www.indeed.com",
        "priority": "u=1, i",
        "referer": "https://www.indeed.com/",
        "sec-ch-ua": '"Chromium";v="124", "Google Chrome";v="124", "Not-A.Brand";v="99"',
        "sec-ch-ua-mobile": "?0",
        "sec-ch-ua-platform": '"Windows"',
        "sec-fetch-dest": "empty",
        "sec-fetch-mode": "cors",
        "sec-fetch-site": "same-site",
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36",
        "x-ms-useragent": "azsdk-js-titleWebappAPIBackend/1.0.0-beta.1 core-http/2.3.2 OS/Win32",
    }
    url = f"https://api-title-webapp.indeed.com/_api/salaries/software%20engineer%20intern?uid=1hsedpf36202v001&ctk=1hqdnajhfk2mo801&locale=en-US&country=US&salaryType=DAILY&location={city}--{state}"
    response = requests.get(url, headers=headers)
    salary_data = response.json()["salaries"]["salaries"]
    if "HOURLY" in salary_data:
        salary = salary_data["HOURLY"]["estimatedMedian"]
    else:
        # if not salary, use national avg
        salary = 22
    cache[(city, state)] = salary
    return salary
```

```python
# create a new df field called salary which either has the provided median salary or scrapes it based on the jobs location
import numpy as np

job_salaries = np.array([])
for i, job in job_df.iterrows():
    if job["job_min_salary"] > 0 and job["job_salary_period"] == "MONTH":
        salary = (job["job_min_salary"] + job["job_max_salary"]) / 2
        job_salaries = np.append(job_salaries, int(salary))
    else:
        # if not salary provided, scrape it
        job_salaries = np.append(
            job_salaries, int(scrape_salary(job["job_city"], job["job_state"]))
        )
job_df["job_salary"] = job_salaries
job_df
```

```python
# save to file
job_df.to_csv("jobs_data.csv")
```
