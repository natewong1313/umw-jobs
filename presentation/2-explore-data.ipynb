```python
import requests
from bs4 import BeautifulSoup

response = requests.get("https://meric.mo.gov/data/cost-living-data-series")
```

```python
soup = BeautifulSoup(response.text, "html.parser")
rows = soup.find("table").find("tbody").find_all("tr")
cost_of_living = {}
for row in rows[:-1]:
    cols = row.find_all("td")
    # rank = int(cols[0].text)
    state = states[cols[1].text]
    cost_index = float(cols[2].text)
    cost_of_living[state] = cost_index
cost_of_living
```

```python
state_salaries = (
    job_df.groupby("job_state")["job_salary"]
    .mean()
    .sort_values(ascending=False)
    .to_dict()
)
state_salaries
```

```python
# now, we will find the best states to work in based on salary and cost of living
best_states = {}
for state, salary in state_salaries.items():
    best_states[state] = salary / cost_of_living[state]
best_states = pd.Series(best_states).sort_values(ascending=False)
best_states.head(10)
```
